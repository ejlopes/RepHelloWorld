```python
import boto3
```

```python
def set_aclfullcontrol_to_owner(pBucket,pInfoOwner,pPrefix):

#     pBucket='sofisa-prod-etl-controls'
#     pInfoOwner='pod.sofisa'
#     pPrefix = 'tb_'



    s3 = boto3.resource('s3')
    s3acl=boto3.client('s3') 
    bucket=s3.Bucket(pBucket)

    vetorKey = []

    for obj in bucket.objects.filter(Prefix=pPrefix):
        if obj.owner['DisplayName'] == pInfoOwner:
            vetorKey.append(obj.key)
   
    for i in range(len(vetorKey)):
        boto3.resource('s3').ObjectAcl(pBucket,vetorKey[i]).put(ACL='bucket-owner-full-control')
    
    return len(vetorKey)
```

```python
a=set_aclfullcontrol_to_owner('sofisa-prod-etl-controls','pod.sofisa','tb_')
b=set_aclfullcontrol_to_owner('sofisa-prod-stage-01','pod.sofisa','tb_')
c=set_aclfullcontrol_to_owner('sofisa-prod-books','pod.sofisa','tb_')

print(a)
print(b)
print(c)
```
