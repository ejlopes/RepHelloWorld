# Identifica informação mais recente de atendimento

```pyspark3
agora=datetime.now()
vdth_proc=agora.strftime("%Y-%m-%d %H:%M:%S")

proc_mais_rec_dl =spark.sql(    
        """
        SELECT
            '{ts}' as dth_proc_dl,
            codEMS_dl,
            substr(mnemonico_exame_dl,locate('-',mnemonico_exame_dl)+1,length(mnemonico_exame_dl)) as exame_trat_dl,
            anomes_dl,
            max(anomesarquivo) as safra_mais_recente_dl
        FROM
            datalake
        --where
        --    data_visita_dl between cast('2019-02-01' as date) and cast('2019-02-28' as date)
        group by 
            1,2,3,4
        """.format(ts=vdth_proc))
proc_mais_rec_dl.registerTempTable("proc_mais_rec_dl")
proc_mais_rec_dl.persist()
proc_mais_rec_dl.count()
```

# Identifica informação mais recente de faturamento

```pyspark3
agora=datetime.now()
vdth_proc=agora.strftime("%Y-%m-%d %H:%M:%S")

proc_mais_rec_fat =spark.sql(    
        """
        SELECT
            '{ts}' as dth_proc_fat,
            codEMS_fat,
            substr(exame_fat,locate('-',exame_fat)+1,length(exame_fat)) as exame_trat_fat,
            anomes_fat,
            max(anomesarquivo) as safra_mais_recente_fat
        FROM
            faturamento
        --where
        --    Dt_Atend_Fixa_fat between cast('2019-02-01' as date) and cast('2019-02-28' as date)
        group by 
            1,2,3,4
        """.format(ts=vdth_proc))
proc_mais_rec_fat.registerTempTable("proc_mais_rec_fat")
proc_mais_rec_fat.persist()
proc_mais_rec_fat.count()
```

# Agrupa dados do exame/atendimento por mês de processamento

```pyspark3
agrup_exame_dl =spark.sql(    
        """
        select
            codEMS_dl,
            substr(mnemonico_exame_dl,locate('-',mnemonico_exame_dl)+1,length(mnemonico_exame_dl)) as exame_trat_dl,
            anomes_dl,
            anomesarquivo as anomesarquivo_dl,
            max(mnemonico_exame_dl) as mnemonico_exame_dl,
            max(codigo_marca_dl) as codigo_marca_dl,  
            max(nome_marca_dl) as nome_marca_dl,
            max(codigo_unidade_dl) as codigo_unidade_dl,
            max(nome_unidade_dl) as nome_unidade_dl,
            max(codigo_convenio_dl) as codigo_convenio_dl,
            max(data_visita_dl) as data_visita_dl,
            max(nome_setor_dl) as nome_setor_dl,
            max(nome_exame_dl) as nome_exame_dl,
            sum(valor_exame_dl) as valor_exame_dl,
            sum(valor_material_medico_dl) as valor_material_medico_dl,
            count(distinct substr(mnemonico_exame_dl,locate('-',mnemonico_exame_dl)+1,length(mnemonico_exame_dl)) ) as qtd_dist_dl,
            count(*) qtd_dl
        FROM
            datalake as a
        --where
        --    data_visita_dl between cast('2019-02-01' as date) and cast('2019-02-28' as date)
        group by 
            1,2,3,4
        """)
agrup_exame_dl.registerTempTable("agrup_exame_dl")
agrup_exame_dl.persist()
agrup_exame_dl.count()
```

# Agrupa dados do exame/faturamento por mês de processamento

```pyspark3
agrup_exame_fat =spark.sql(    
        """
        SELECT
            codEMS_fat,
            substr(exame_fat,locate('-',exame_fat)+1,length(exame_fat)) as exame_trat_fat,
            anomes_fat,
            anomesarquivo as anomesarquivo_fat,
            max(exame_fat) as exame_fat,
            max(Unid_fat) Unid_fat,
            max(ConvenioEMS_fat) as ConvenioEMS_fat,
            max(Convenio_fat) as Convenio_fat,
            max(CodAMB_fat) as CodAMB_fat,
            max(Pendente_fat) as Pendente_fat,
            sum(ValorTotalExame_fat) as ValorTotalExame_fat,
            sum(ValorTotalExameRec_fat) as ValorTotalExameRec_fat,
            sum(ValorMaterial_fat) as ValorMaterial_fat,
            sum(ValorMedicamento_fat) as ValorMedicamento_fat,
            sum(ValorFilme_fat) as ValorFilme_fat,
            sum(ValorTaxas_fat) as ValorTaxas_fat,
            sum(ValorExame_fat) as ValorExame_fat,
            sum(ValorExameV1_CH_fat) as ValorExameV1_CH_fat,
            sum(Valor_fat) as Valor_fat,
            max(CodStatus_fat) as CodStatus_fat,
            count(distinct substr(exame_fat,locate('-',exame_fat)+1,length(exame_fat))) as qtd_dist_fat,
            count(*) as qtd_fat
        FROM
            faturamento
        --where
        --    Dt_Atend_Fixa_fat between cast('2019-02-01' as date) and cast('2019-02-28' as date)
        group by 
            1,2,3,4
        """)
agrup_exame_fat.registerTempTable("agrup_exame_fat")
agrup_exame_fat.persist()
agrup_exame_fat.count()
```

# Filtra atendimento mais recente para relatório

```pyspark3
exame_mais_rec_dl =spark.sql(    
        """
        select
            b.dth_proc_dl,
            a.*
        FROM
                       agrup_exame_dl as a
            inner join proc_mais_rec_dl as b
        on 
            (
                    a.codEMS_dl = b.codEMS_dl 
                and a.exame_trat_dl = b.exame_trat_dl 
                and a.anomes_dl = b.anomes_dl
                and a.anomesarquivo_dl = b.safra_mais_recente_dl
            )

        """)
exame_mais_rec_dl.registerTempTable("exame_mais_rec_dl")
exame_mais_rec_dl.persist()
exame_mais_rec_dl.count()
```

# Filtra faturamento mais recente para relatório

```pyspark3
exame_mais_rec_fat =spark.sql(    
        """
        select
            b.dth_proc_fat,
            a.*
        FROM
                       agrup_exame_fat as a
            inner join proc_mais_rec_fat as b
        on 
            (
                    a.codEMS_fat = b.codEMS_fat
                and a.exame_trat_fat = b.exame_trat_fat
                and a.anomes_fat = b.anomes_fat
                and a.anomesarquivo_fat = b.safra_mais_recente_fat
            )

        """)
exame_mais_rec_fat.registerTempTable("exame_mais_rec_fat")
exame_mais_rec_fat.persist()
exame_mais_rec_fat.count()
```

# # Join de conciliação Atendido x Faturado

```pyspark3
agora=datetime.now()
vdth_proc=agora.strftime("%Y-%m-%d %H:%M:%S")

full_join_exame_dlxfat =spark.sql(    
        """
        SELECT
            1 as qtd_exame_conciliacao,
            current_date as dt_proc_conciliacao,
            '{ts}' as dth_proc_conciliacao,
            '{ts}' as dth_proc_conciliacao_partition,
            coalesce(a.anomes_dl,f.anomes_fat) as ref_anomes,
            coalesce(a.anomes_dl,f.anomes_fat) as ref_anomes_partition,
            case 
                when a.codEMS_dl is not null and f.CodEMS_fat is not null then 'Atendido & Faturado'
                when a.codEMS_dl is not null and f.CodEMS_fat is null then 'Somente Atendido'
                when a.codEMS_dl is null and f.CodEMS_fat is not null then 'Somente Faturado' 
                else 'Erro'
            end as Flag_Status_Conciliacao, 
            a.*,
            f.*
        FROM
                  exame_mais_rec_dl as a
        FULL JOIN exame_mais_rec_fat as f
        ON 
            (
                    a.codEMS_dl = f.CodEMS_fat 
                and a.exame_trat_dl = f.exame_trat_fat
                and a.anomes_dl = f.anomes_fat
            )
        """.format(ts=vdth_proc)
        )
full_join_exame_dlxfat.registerTempTable("full_join_exame_dlxfat")
full_join_exame_dlxfat.persist()
full_join_exame_dlxfat.count()
```
